<h4>Checking no. of attacks with respect to year</h4>

```python
plt.subplots(figsize=(20,10))
sb.countplot('iyear',data=df,palette='tab10')
plt.xticks(rotation=70)
plt.title('no. of terror attacks')
plt.show()
```

<h4>No. of attacks in 2017 and 1970</h4>

```python
yr=df.iyear.value_counts().to_dict()
print('No.of attacks in year 2017 is',yr[2017],'\n','and no. of attacks in year 1970 is',yr[1970])
```

* <b>From this we can say that there has been a significant increase in no. of attacks from 1970 to 2017</b>

<h4>Region wise attack count</h4>

```python
df['region_txt'].value_counts()
plt.figure(figsize=(15,8))
sb.countplot(x='region_txt',data=df)
plt.xticks(rotation=70)
plt.show()
```

* <b>from the abobe graph we observed that Middle East & North Africa and South Asia is most attacked regions</b>

<h4>Top countries affected</h4>

```python
df['country'].value_counts()
plt.figure(figsize=(20,10))
sb.barplot(x=df['country_txt'].value_counts()[:20].index,y=df['country_txt'].value_counts()[:20].values)
plt.title('Top Countries Affected')
plt.xticks(rotation=90)
plt.xlabel('Countries')
plt.ylabel('No. of terror attack')
plt.show()
```

* <b>Iraq is most attacked country</b>

```python
df['gname'].value_counts()[:20]
```

```python
plt.figure(figsize=(15,6))
sb.barplot(x=df['gname'].value_counts()[:20].index,y=df['gname'].value_counts()[:20].values)
plt.xticks(rotation=90)
plt.title('Total number of attacks group-wise')
plt.xlabel('Country')
plt.ylabel('Total attack counts')
plt.show()
```

```python
df['attacktype1_txt'].value_counts()
```

```python
plt.figure(figsize=(10,10))
sb.countplot(x=df['attacktype1_txt'],order=df['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.show()
```

* <b>From this we can saty that bombing/Explosion have been used the most i.e. more than 80000 times</b>

<h4>Visualizing the type of people attacked by terror groups</h4>

```python
plt.figure(figsize=(20,10))
sb.countplot(df['targtype1_txt'],order=df['targtype1_txt'].value_counts().index,palette='rocket')
plt.xticks(rotation=90)
plt.show()
```

* <b>By this we observed that Private Citizen &Property has been targeted the most i.e. more tahn 40000 times</b>
